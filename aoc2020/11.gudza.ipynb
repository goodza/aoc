```python
from copy import deepcopy
```

```python
def check(i,j,F):
    other,same = 0,-1
    for k in range(-1,2):
        for kk in range(-1,2):
            if 0<=j+kk<len(F[0]) and 0<=i+k<len(F):          
                if F[i+k][j+kk] not in [F[i][j],'.']:
                    other += 1
                if F[i+k][j+kk] == F[i][j]:
                    same += 1
                    
    return other,same
                
             
def sit(F,first_rule):
    
    FF = deepcopy(F)
    for i in range(0,len(F)):
        for j in range(0,len(F[0])):
            if F[i][j] != '.':
                other,same = check(i,j,F)
                if first_rule and F[i][j] == 'L' and other == 0:
                    FF[i][j] = '#'
                if not first_rule and F[i][j] == '#' and same >= 4:
                    FF[i][j] = 'L'
    return FF, not first_rule


def oc_count(F):
    c = 0
    for i in F:
        for j in i:
            if j == '#':
                c += 1
    return c
```

```python
def first_seat_check(i,j,F):
    occupy = -1

    for k in [-1,0,1]:
        for kk in [-1,0,1]:
        
            for n in range(1,max(len(F[0])-j,j,i,len(F)-i)+1):
                
                if 0<=j+kk*n<len(F[0]) and 0<=i+k*n<len(F):  
                    if F[i+k*n][j+kk*n] == 'L':
#                         if F[i][j] == 'L':
                        break
#                         if F[i][j] == '#':
#                             continue
                    if F[i+k*n][j+kk*n] == '#':
                        if F[i][j] == 'L':
                            return False,occupy
                        if F[i][j] == '#':
                            occupy += 1
                            break
                        
    return True,occupy


def adv_sit(F,first_rule):
    
    FF = deepcopy(F)
    for i in range(0,len(F)):
        for j in range(0,len(F[0])):
            if F[i][j] != '.':
                free,occupy = first_seat_check(i,j,F)
                if first_rule and F[i][j] == 'L' and free:
#                     print(i,j,occupy)
                    FF[i][j] = '#'
                if not first_rule and F[i][j] == '#' and occupy >= 5:
                    FF[i][j] = 'L'
    return FF, not first_rule
```

```python
def cycle(F):
    first_rule = True
    F_pre = F
    while True:
        F_next,first_rule = adv_sit(F_pre,first_rule)
        if F_next == F_pre:
            print(oc_count(F_next),oc_count(F_pre))
            return F_next
        F_pre = deepcopy(F_next)
        del F_next

cycle(f)
```

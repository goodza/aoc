```python
import re
```

```python
f = open('4.txt','r+').read().split("\n\n")
```

```python
fields = {'hcl', 'ecl', 'byr', 'iyr', 'eyr', 'hgt', 'pid', 'cid'}

c = 0

for i in f:
    j = re.split('[ \n]', i) #split by space either newline
#     print(j)
    t = {'cid'}
    for jj in j:
        t.add(jj.split(':')[0])
    print(t)
    if t == fields:
        c += 1

print(c)
```

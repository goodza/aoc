```python
def Points(w):
    P = []
    L = []
    x = 0
    y = 0
    for i in w:
        c = i[0]
        o = int(i[1:len(i)])
        if c == 'R':
            x = x + o
        if c == 'L':
            x = x - o
        if c == 'U':
            y = y + o
        if c == 'D':
            y = y - o
        P.append([x,y])
        L.append(o)
    return P, L
```

```python
p1, l1 = Points(w1)
print(l1)
p2, l2 = Points(w2)
print(l2)
```

```python
#find intersections

S = []
x1 = 0
y1 = 0
la = 0
ai = 0
SL = []

for a in p1:
    
    x2 = a[0]
    y2 = a[1]
    x3 = 0
    y3 = 0
    lb = 0
    bi = 0
    la += l1[ai]
    
   
    for b in p2: 
        
        x4 = b[0]
        y4 = b[1]
        d = (x4-x3)*(y1-y2)-(x1-x2)*(y4-y3)
        lb += l2[bi]
        
        if d != 0:
            ta = ((y3-y4)*(x1-x3)+(x4-x3)*(y1-y3))/d
            tb = ((y1-y2)*(x1-x3)+(x2-x1)*(y1-y3))/d
            #intersected!
            if 0<=ta<=1 and 0<=tb<=1:
                xs = int((x2-x1)*ta+x1)
                ys = int((y2-y1)*ta+y1)
                S.append([xs,ys])
                SL.append(lb-l2[bi]+abs(xs-x3)+abs(ys-y3)+abs(xs-x1)+abs(ys-y1)+la-l1[ai])
                
        x3 = x4
        y3 = y4
        bi += 1
      
    x1 = x2
    y1 = y2
        
    ai += 1
    
S = S[1:len(S)]
print(len(S))

print(len(SL))

print(min(SL))
```
